```python
from textblob import TextBlob
from transformers import pipeline
```

```python
review_output_file = '..\..\yelp_dataset\medium\processed_review_sentiment.csv'
```

```python
# Function to determine sentiment
def get_sentiment(text):
    analysis = TextBlob(text)
    # Classify as positive if sentiment polarity is above 0, otherwise negative
    return 'Positive' if analysis.sentiment.polarity > 0 else 'Negative'
```

```python
# Initialize the sentiment analysis pipeline using a pre-trained model
sentiment_analyzer = pipeline('sentiment-analysis')
```

```python
# Function to get the sentiment from the model
def get_sentiment2(text):
    result = sentiment_analyzer(text)[0]
    # Classify as positive if label is 'POSITIVE', otherwise negative
    return 'Positive' if result['label'] == 'POSITIVE' else 'Negative'
```

```python
# Apply sentiment analysis to the 'text' column
review_df['sentiment'] = review_df['text'].apply(get_sentiment)

# Save the processed data with sentiment to a new CSV file
review_df.to_csv(review_output_file, index=False)
```

#### create new category from business dataset

```python

category_output_file = '..\..\yelp_dataset\medium\category.csv'
business_category_output_file = '..\..\yelp_dataset\medium\category_business.csv'
```

```python

# Split the 'categories' column into individual categories, drop NaNs, trim spaces, and explode into separate rows
categories = business_df['categories'].dropna().str.split(',').explode().str.strip()
```

```python
# Get unique categories and create a category_id
distinct_categories = categories.drop_duplicates().reset_index(drop=True).reset_index()
distinct_categories.columns = ['category_id', 'category']
distinct_categories['category_id'] += 1  # Start IDs from 1
```

```python

# Merge back to create a mapping of business_id to category_id
business_categories = business_df[['business_id']].join(categories)
business_categories = business_categories.merge(distinct_categories, left_on='categories', right_on='category')
business_categories = business_categories[['business_id', 'category_id']]
```

```python

# Save the distinct categories to 'category.csv'
distinct_categories.to_csv(category_output_file, index=False)
```

```python
# Save the business to category mapping to 'business_category.csv'
business_categories.to_csv(business_category_output_file, index=False)
```
